```python
import pandas as pd
import numpy as np
```

```python
df_gdp = pd.read_csv('global-gdppercapita-byYear.csv')
df_gdp
```

```python
gdp_list = np.array(df_gdp.apply(lambda row: [row['Country Name'], 2000, row['2000 [YR2000]']], axis = 1))
gdp_list = np.hstack((gdp_list, df_gdp.apply(lambda row: [row['Country Name'], 2005, row['2005 [YR2005]']], axis = 1)))
gdp_list = np.hstack((gdp_list, df_gdp.apply(lambda row: [row['Country Name'], 2010, row['2010 [YR2010]']], axis = 1)))
gdp_list = np.hstack((gdp_list, df_gdp.apply(lambda row: [row['Country Name'], 2013, row['2013 [YR2013]']], axis = 1)))
gdp_list = np.hstack((gdp_list, df_gdp.apply(lambda row: [row['Country Name'], 2015, row['2015 [YR2015]']], axis = 1)))
```

```python
gdp_df = pd.DataFrame(gdp_list, columns = ['a'])
```

```python
gdp_df['Country'] = gdp_df['a'].apply(lambda x: x[0])
gdp_df['Year'] = gdp_df['a'].apply(lambda x: x[1])
gdp_df['GDPperCapita'] = gdp_df['a'].apply(lambda x: x[2])
```
